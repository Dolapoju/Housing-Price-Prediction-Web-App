## House Price Prediction

#### imports

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np 
import seaborn as sns
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression, Lasso,Ridge
import pickle
```

### fetching Data and details about how it was collected

```python
housing = fetch_california_housing(as_frame=True)
df = housing.frame
```

```python
print(housing.DESCR)
```

The MedHouseVal is of a scale of 1 to $100000
from the decription we know that there are no missing values, but we will confirm this

### Exploratory Data Analysis

```python
df.head()
```

```python
sns.heatmap(df.isnull(),cmap="coolwarm",cbar=False) # using a heatmap to check for missing values,cmap="coolawarm" makes blue 0 and red 1
```

The MedHouseVal is the price which we will be tring to predict

From the description the reason features are averages is because the data was gotten from blocks of houses not from each house

### Choosing Features

```python
corr=df.corr()["MedHouseVal"]
corr.sort_values(ascending=False).head(9)# there are only 9 features
```

Attempt to create features with a better correlation

```python
df["population_per_bedroom"] = df["Population"] / df["AveBedrms"]
corr=df.corr()["MedHouseVal"]
corr.sort_values(ascending=False).head(10)
```

Conclusion is that we already have the best features

```python
df = df.drop(columns=["population_per_bedroom"])
```

```python
chosen_features=[ "MedInc","AveRooms", "HouseAge","AveBedrms","Longitude","Latitude"]
x=df[chosen_features]
y=df["MedHouseVal"]
```

### Linear Regression

```python
scaler = StandardScaler()
scaled_x = scaler.fit_transform(x)
linearmodel = LinearRegression()
linearmodel.fit(scaled_x, y)
```

### Ridge Model

```python
ridge_pipeline = Pipeline([
    ('model', Ridge())
])
ridgehyperparams = {'model__alpha': [0.1, 1, 10, 50, 100]}
ridge_grid = GridSearchCV(ridge_pipeline,ridgehyperparams,cv=5,scoring='neg_root_mean_squared_error')
ridge_grid.fit(scaled_x, y)
```

### LASSO Model

```python
lasso = Pipeline(steps=[
    ('model', Lasso())
])
lasso_params = { 'model__alpha': [0.0005, 0.001, 0.01, 0.1, 1, 10] }
lasso_grid = GridSearchCV(lasso, lasso_params, cv=5, scoring='neg_root_mean_squared_error')
lasso_grid.fit(scaled_x, y)
```

### Selecting the Best Model

```python
def evaluate_model(model, X, y):
    predictions = model.predict(X)# this is getting the predicted prices
    # there are about 5 ways to evaluate a model, we will be using R (R-squared)
    r2 = r2_score(y, predictions)
    return r2
```

```python
tabledata = {
    "Linear regression": evaluate_model(linearmodel, scaled_x, y),
    "Ridge": evaluate_model(ridge_grid.best_estimator_, scaled_x, y),
    "LASSO": evaluate_model(lasso_grid.best_estimator_, scaled_x, y)
}

pd.DataFrame(tabledata, index=["R2 values"])
```

```python
models_list = {
    "Linear regression": linearmodel,
    "Ridge": ridge_grid.best_estimator_,
    "LASSO": lasso_grid.best_estimator_
}

name = max(tabledata, key=lambda m: tabledata[m])
best_model = models_list[name]

print("Best R value:", tabledata[name])
print("Best Model:", name)
```

### Writing into a file with Pickle

```python
with open("best_model.pkl", "wb") as file:
    pickle.dump(best_model, file)
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
```
